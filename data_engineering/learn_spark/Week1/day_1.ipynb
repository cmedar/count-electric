## 1. Why Spark?

Spark is a **distributed compute engine** — you write Python, Spark runs it across a cluster of machines in parallel.

| | Hadoop MapReduce | Apache Spark |
|---|---|---|
| Processing | Disk-based (writes between steps) | In-memory (keeps data in RAM) |
| Speed | Slow (disk I/O) | Up to 100× faster for iterative jobs |
| API | Low-level Java/Python wrappers | High-level: DataFrame, SQL, ML, Streaming |
| Use cases | Batch only | Batch + streaming + ML + graph |
| Fault tolerance | HDFS replication | RDD lineage |

> **Key insight:** Spark doesn't replace Hadoop's storage (HDFS/S3) — it replaces MapReduce as the compute layer.

## 2. SparkSession

The `SparkSession` is your single entry point to all Spark functionality.

```
SparkSession
    ├── SparkContext   (low-level RDD API)
    ├── SQLContext     (Spark SQL)
    └── HiveContext    (Hive integration)
```

In **Databricks**, `spark` is already created for you — no setup needed.  
Outside Databricks (local or custom cluster), you build it:

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("MyApp").getOrCreate()
```

````python
# Confirm Spark is available and check version
print(f"Spark version : {spark.version}")
print(f"App name      : {spark.sparkContext.appName}")
print(f"Master        : {spark.sparkContext.master}")
````

## 3. Creating a DataFrame

A **DataFrame** is a distributed table — rows and named, typed columns.  
Think of it as a pandas DataFrame, but split across many machines.

````python
data = [
    ("Alice",   34, "Engineering", 95000),
    ("Bob",     28, "Marketing",   72000),
    ("Carol",   41, "Engineering", 110000),
    ("Dave",    25, "Sales",        58000),
    ("Eve",     36, "Marketing",    81000),
    ("Frank",   29, "Engineering",  88000),
]
columns = ["name", "age", "department", "salary"]

df = spark.createDataFrame(data, columns)
df.show()
````

## 4. Inspecting a DataFrame

````python
# Schema — column names and inferred types
df.printSchema()
````

````python
# Row count — this triggers an action (more on Day 3)
print(f"Row count: {df.count()}")

# Column names and dtypes
print(df.columns)
print(df.dtypes)
````

## 5. Basic Operations: select() and filter()

````python
# select() — choose columns (like SQL SELECT)
df.select("name", "department").show()

# filter() — keep rows matching a condition (like SQL WHERE)
df.filter(df.age > 30).show()

# Chain them together
df.select("name", "salary").filter(df.department == "Engineering").show()
````

## 6. The Evolution: RDD → DataFrame → Dataset

| Layer | What it is | Used today? |
|---|---|---|
| **RDD** | Raw distributed collection, no schema, low-level | Rarely — only for niche control |
| **DataFrame** | Distributed table with schema — your main tool | Yes |
| **Dataset** | Typed DataFrame (Scala/Java only) | Not in Python |

In Python you work exclusively with **DataFrames**. The exam tests DataFrame API deeply.

> **Exam tip:** `DataFrame` in Python is actually `Dataset[Row]` under the hood. You may see this in exam questions.

````python
# A DataFrame is a Dataset[Row] internally
print(type(df))  # pyspark.sql.dataframe.DataFrame

# Access the underlying SparkContext via the session
sc = spark.sparkContext
print(f"Default parallelism: {sc.defaultParallelism}")
````
